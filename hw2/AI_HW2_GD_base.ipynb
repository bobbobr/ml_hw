```python
import sympy as sp

x = sp.Symbol('x')
f = x**2 - 30*x + 1
g = -x**3 - 7*x**2 + 5*x + 17
h = x**2 - 1/x
j = 1/x**5 + x**17 + 8

f_prime = sp.diff(f, x)
g_prime = sp.diff(g, x)
h_prime = sp.diff(h, x)
j_prime = sp.diff(j, x)

f_crit_points = [p.evalf() for p in sp.solveset(f_prime, x, domain=sp.S.Reals)]
g_crit_points = [p.evalf() for p in sp.solveset(g_prime, x, domain=sp.S.Reals)]
h_crit_points = [p.evalf() for p in sp.solveset(h_prime, x, domain=sp.S.Reals)]
j_crit_points = [p.evalf() for p in sp.solveset(j_prime, x, domain=sp.S.Reals)]

f_double_prime = sp.diff(f_prime, x)
g_double_prime = sp.diff(g_prime, x)
h_double_prime = sp.diff(h_prime, x)
j_double_prime = sp.diff(j_prime, x)

f_results = [(point, f_double_prime.subs(x, point).evalf(), f.subs(x, point).evalf()) for point in f_crit_points]
g_results = [(point, g_double_prime.subs(x, point).evalf(), g.subs(x, point).evalf()) for point in g_crit_points]
h_results = [(point, h_double_prime.subs(x, point).evalf(), h.subs(x, point).evalf()) for point in h_crit_points]
j_results = [(point, j_double_prime.subs(x, point).evalf(), j.subs(x, point).evalf()) for point in j_crit_points]

print("Results for f(x):")
for point, second_derivative, value in f_results:
    extrema_type = "Minimum" if second_derivative > 0 else "Maximum" if second_derivative < 0 else "Inflection Point"
    print(f"x = {point}, f(x) = {value}, Type: {extrema_type}")

print("\nResults for g(x):")
for point, second_derivative, value in g_results:
    extrema_type = "Minimum" if second_derivative > 0 else "Maximum" if second_derivative < 0 else "Inflection Point"
    print(f"x = {point}, g(x) = {value}, Type: {extrema_type}")

print("\nResults for h(x):")
for point, second_derivative, value in h_results:
    extrema_type = "Minimum" if second_derivative > 0 else "Maximum" if second_derivative < 0 else "Inflection Point"
    print(f"x = {point}, h(x) = {value}, Type: {extrema_type}")

print("\nResults for j(x):")
for point, second_derivative, value in j_results:
    extrema_type = "Minimum" if second_derivative > 0 else "Maximum" if second_derivative < 0 else "Inflection Point"
    print(f"x = {point}, j(x) = {value}, Type: {extrema_type}")
```

```python
# your code here

import sympy as sp

def find_inflection_points(function):
    x = sp.Symbol('x')
    
    first_derivative = sp.diff(function, x)
    second_derivative = sp.diff(first_derivative, x)
    
    candidate_points = sp.solveset(second_derivative, x, domain=sp.S.Reals)
    
    inflection_points = []
    for point in candidate_points:
        left_limit = second_derivative.subs(x, point - 0.01).evalf()
        right_limit = second_derivative.subs(x, point + 0.01).evalf()

        if left_limit * right_limit < 0: 
            inflection_points.append(point.evalf())

    return inflection_points

x = sp.Symbol('x')
f = x**2 - 30*x + 1
g = -x**3 - 7*x**2 + 5*x + 17
h = x**2 - 1/x
j = 1/x**5 + x**17 + 8

inflection_f = find_inflection_points(f)
inflection_g = find_inflection_points(g)
inflection_h = find_inflection_points(h)
inflection_j = find_inflection_points(j)

print("Inflection points for f(x):", inflection_f)
print("Inflection points for g(x):", inflection_g)
print("Inflection points for h(x):", inflection_h)
print("Inflection points for j(x):", inflection_j)
```

## Задание 6 (1 балл)

Модифицируйте функцию, осуществляющую метод градиентного спуска с семинара: добавьте опцию останавливать метод, если абсолютное значение производной функции в точке < некоторого $\delta$ (delta добавьте в аргументы функции).

```python
# your code here


def gradient_descent(func, x_start, learning_rate, epsilon, num_iterations, detla):
    x_curr = x_start
    df_x = sp.diff(func(x)) 

    trace = []
    trace.append(x_curr)

    for i in range(num_iterations):
      grad = df_x.subs(x, x_curr) 
      
      if abs(grad) < detla:
            print(f"Остановлено по критерию delta на {i}-й итерации")
            return x_curr, trace
      x_new = x_curr - learning_rate * df_x.subs(x, x_curr)
      trace.append(x_new)

      if abs(x_new - x_curr) < epsilon:
        return x_curr, trace

      x_curr = x_new

    return float(x_curr), trace
```

```python
def gradient_descent_for_j(f, x_start, learning_rate, epsilon, num_iterations, delta):
    x = sp.Symbol('x')
    x_curr = x_start
    trace = [x_curr]
    df_x = sp.diff(f, x)  

    for i in range(num_iterations):
        grad_value = df_x.subs(x, x_curr)

        if not isinstance(grad_value, (int, float, sp.Float)):
            break

        grad_value = float(grad_value)  

        if abs(grad_value) < delta:
            print(f"Остановлено на итерации {i}: abs(grad) < delta")
            return x_curr, trace

        adaptive_lr = learning_rate / (1 + i * 0.01)  
        x_new = x_curr - adaptive_lr * grad_value

        if x_new <= 0:
            print(f"x вышел за границ.")
            break

        trace.append(float(x_new))

        if abs(x_new - x_curr) < epsilon:
            return x_new, trace

        x_curr = x_new

    print("Достигнуто максимальное число итераций")
    return x_curr, trace

def plot_function_and_minimum(f, x_range, x_min, f_min, title):
    x_vals = np.linspace(*x_range, 500)
    f_lambdified = sp.lambdify(sp.Symbol('x'), f, 'numpy')
    y_vals = f_lambdified(x_vals)

    plt.figure(figsize=(8, 6))
    plt.plot(x_vals, y_vals, label=f"{title}(x)", color="blue")
    plt.scatter([x_min], [f_min], color="red", label="Найденный минимум")
    plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
    plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
    plt.annotate(f"({x_min:.2f}, {f_min:.2f})", (x_min, f_min), 
                 textcoords="offset points", xytext=(-30, 10), ha="center")
    plt.title(f"Gradient Descent on {title}(x)")
    plt.xlabel("x")
    plt.ylabel(f"{title}(x)")
    plt.legend()
    plt.grid()
    plt.show()

x = sp.Symbol('x')
j_expr = sp.sympify("1/x**5 + x**17 + 8")

learning_rate = 0.01
epsilon = 1e-6
num_iterations = 1000
delta = 1e-6

best_min = float('inf')
best_x0 = None

for _ in range(10):
    x0 = np.random.uniform(0.5, 2)  
    print(f"Начальная точка: {x0}")

    min_val, trace = gradient_descent_for_j(j_expr, x0, learning_rate, epsilon, num_iterations, delta)

    if isinstance(min_val, (int, float)) and min_val < best_min:
        best_min = min_val
        best_x0 = x0

print(f"Лучшее начальное приближение (x0): {best_x0}")
if isinstance(best_min, (int, float)) and np.isfinite(best_min):
    print(f"Функция: j(x)")
    print(f"Начальная точка (x_0): {best_x0}")
    print(f"Найденный минимум: {best_min}")
else:
    print("Найденный минимум не является числом.")

plot_function_and_minimum(j_expr, (0.5, 2), best_min, j_expr.subs(x, best_min), "j")
```

## Задание 8 (1.5 балла)

Реализуйте [метод Ньютона](https://en.wikipedia.org/wiki/Newton%27s_method_in_optimization) (метод второго порядка) для поиска минимума функции.

Про метод ещё можно почитать [здесь](https://math.semestr.ru/optim/minimum.php).

```python
# your code here

import sympy as sp

def newton_method(func, x_start, epsilon=1e-6, max_iterations=100, delta=1e-6):
    x = sp.Symbol('x')
    d_x = sp.diff(func, x)  
    d_xx = sp.diff(d_x, x)  

    x_curr = x_start
    trace = [x_curr]

    for i in range(max_iterations):
        grad = d_x.subs(x, x_curr)  
        hessian = d_xx.subs(x, x_curr)  

        if abs(grad) < delta:
            print(f"Градиент близок к 0 на {i}-й итерации.")
            break

        if hessian == 0:
            raise ValueError("Гессиан равен 0. Метод Ньютона не применим.")

        x_new = x_curr - grad / hessian  
        trace.append(x_new)

        if abs(x_new - x_curr) < epsilon:
            print(f"Метод сошелся на {i}-й итерации.")
            break

        x_curr = x_new

    return x_curr, trace
```

```python
def gradient_descent_for_10(func, x_start, learning_rate, epsilon, num_iterations, detla):
    x_curr = x_start
    df_x = sp.diff(func,x) 

    trace = []
    trace.append(x_curr)

    for i in range(num_iterations):
      grad = df_x.subs(x, x_curr) 
      
      if abs(grad) < detla:
            print(f"Остановлено по критерию delta на {i}-й итерации")
            return float(x_curr), len(trace), trace
      x_new = x_curr - learning_rate * df_x.subs(x, x_curr)
      trace.append(x_new)

      if abs(x_new - x_curr) < epsilon:
        return float(x_curr), len(trace), trace

      x_curr = x_new

    return float(x_curr), len(trace), trace
```

```python
def newton_method_for_10(func, x_start, epsilon=1e-6, max_iterations=100, delta=1e-6):
    x = sp.Symbol('x')
    d_x = sp.diff(func, x)
    d_xx = sp.diff(d_x, x)  

    x_curr = float(x_start)
    trace = [x_curr]

    for i in range(max_iterations):
        grad = d_x.subs(x, x_curr)  
        hessian = d_xx.subs(x, x_curr)  

        if abs(grad) < delta:
            print(f"Остановлено на итерации {i}: abs(grad) < delta")
            return x_curr, i + 1, trace

        if abs(hessian) < epsilon:
            print(f"Гессиан слишком мал на {i}-й итерации. Метод Ньютона не применим.")
            break

        x_new = x_curr - grad / hessian  
        trace.append(x_new)

        if abs(x_new - x_curr) < epsilon:
            print(f"Метод сошелся на {i}-й итерации.")
            return x_new, i + 1, trace

        x_curr = x_new

    print("Достигнуто максимальное число итераций.")
    return x_curr, max_iterations, trace
```

```python
# your code here

x = sp.Symbol('x')
functions = {
    "f": x**2 - 30*x + 1,
    "g": -x**3 - 7*x**2 + 5*x + 17,
    "h": x**2 - 1/x,
    "j": 1/x**5 + x**17 + 8
}
initial_points = {"f": 0, "g": -6, "h": 0.5, "j": 0.2}
true_minima = {"f": 15, "g": -5, "h": -((4)**(1/3))/2, "j": (5/17)**(1/22)}  
```

```python
# Сравнение методов

# Фиксированное кол-во 
learning_rate = 0.01
epsilon = 1e-5
delta = 1e-6
max_iterations = 1000
N = 10

gd_results = []
newton_results = []
accuracy_comparison = []

for name, f_expr in functions.items():
    x_start = initial_points[name]
    print(f"Функция: {name}(x)")
    
    gd_min, gd_iters, gd_trace = gradient_descent_for_10(f_expr, x_start, learning_rate, epsilon, max_iterations, delta)
    gd_error = abs(gd_min - true_minima[name]) if true_minima[name] is not None else "N/A"

    try:
        newton_min, newton_iters, newton_trace = newton_method_for_10(f_expr, x_start, epsilon, max_iterations)
        newton_error = abs(newton_min - true_minima[name]) if true_minima[name] is not None else "N/A"
    except Exception as e:
        newton_min, newton_iters, newton_trace = None, None, []
        newton_error = str(e)

    gd_n_iters_min = gd_trace[N-1] if len(gd_trace) >= N else gd_trace[-1]
    newton_n_iters_min = (
        newton_trace[N-1] if len(newton_trace) >= N and len(newton_trace) > 0 
        else newton_trace[-1] if len(newton_trace) > 0 
        else None
    )

    accuracy_comparison.append((name, gd_n_iters_min, newton_n_iters_min))

    gd_results.append((name, gd_iters, gd_min, gd_error))
    newton_results.append((name, newton_iters, newton_min, newton_error))

print("\nСравнение точности при N=10 итерациях:")
for comp in accuracy_comparison:
    print(f"Функция: {comp[0]} | Градиентный спуск: {comp[1]} | Метод Ньютона: {comp[2]}")
```

Реализуйте оптимизацию метода градиентного спуска - Momentum. Необходимые выкладки вы можете подсмотреть [здесь](https://education.yandex.ru/handbook/ml/article/optimizaciya-v-ml).

```python
# your code here

def gradient_descent_momentum(f_prime, x_start, learning_rate, momentum, epsilon, num_iterations):
    x_curr = x_start
    v = 0  
    trace = [x_curr]  

    for i in range(num_iterations):
        grad = f_prime(x_curr)
        v = momentum * v - learning_rate * grad  
        x_new = x_curr + v  
        trace.append(x_new)

        if abs(x_new - x_curr) < epsilon:  
            break

        x_curr = x_new

    return x_curr, trace
```
