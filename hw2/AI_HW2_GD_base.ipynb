## Задание 2 (1 балл)

Найдите экстремумы этих функций в Python.

```python
import sympy as sp

x = sp.Symbol('x')
f = x**2 - 30*x + 1
g = -x**3 - 7*x**2 + 5*x + 17
h = x**2 - 1/x
j = 1/x**5 + x**17 + 8

f_prime = sp.diff(f, x)
g_prime = sp.diff(g, x)
h_prime = sp.diff(h, x)
j_prime = sp.diff(j, x)

f_crit_points = [p.evalf() for p in sp.solveset(f_prime, x, domain=sp.S.Reals)]
g_crit_points = [p.evalf() for p in sp.solveset(g_prime, x, domain=sp.S.Reals)]
h_crit_points = [p.evalf() for p in sp.solveset(h_prime, x, domain=sp.S.Reals)]
j_crit_points = [p.evalf() for p in sp.solveset(j_prime, x, domain=sp.S.Reals)]

f_double_prime = sp.diff(f_prime, x)
g_double_prime = sp.diff(g_prime, x)
h_double_prime = sp.diff(h_prime, x)
j_double_prime = sp.diff(j_prime, x)

f_results = [(point, f_double_prime.subs(x, point).evalf(), f.subs(x, point).evalf()) for point in f_crit_points]
g_results = [(point, g_double_prime.subs(x, point).evalf(), g.subs(x, point).evalf()) for point in g_crit_points]
h_results = [(point, h_double_prime.subs(x, point).evalf(), h.subs(x, point).evalf()) for point in h_crit_points]
j_results = [(point, j_double_prime.subs(x, point).evalf(), j.subs(x, point).evalf()) for point in j_crit_points]

print("Results for f(x):")
for point, second_derivative, value in f_results:
    extrema_type = "Minimum" if second_derivative > 0 else "Maximum" if second_derivative < 0 else "Inflection Point"
    print(f"x = {point}, f(x) = {value}, Type: {extrema_type}")

print("\nResults for g(x):")
for point, second_derivative, value in g_results:
    extrema_type = "Minimum" if second_derivative > 0 else "Maximum" if second_derivative < 0 else "Inflection Point"
    print(f"x = {point}, g(x) = {value}, Type: {extrema_type}")

print("\nResults for h(x):")
for point, second_derivative, value in h_results:
    extrema_type = "Minimum" if second_derivative > 0 else "Maximum" if second_derivative < 0 else "Inflection Point"
    print(f"x = {point}, h(x) = {value}, Type: {extrema_type}")

print("\nResults for j(x):")
for point, second_derivative, value in j_results:
    extrema_type = "Minimum" if second_derivative > 0 else "Maximum" if second_derivative < 0 else "Inflection Point"
    print(f"x = {point}, j(x) = {value}, Type: {extrema_type}")
```

## Задание 4 (1 балла)

Найдите точки перегиба этих функций в Python.

```python
# your code here

import sympy as sp

def find_inflection_points(function):
    x = sp.Symbol('x')
    
    first_derivative = sp.diff(function, x)
    second_derivative = sp.diff(first_derivative, x)
    
    candidate_points = sp.solveset(second_derivative, x, domain=sp.S.Reals)
    
    inflection_points = []
    for point in candidate_points:
        left_limit = second_derivative.subs(x, point - 0.01).evalf()
        right_limit = second_derivative.subs(x, point + 0.01).evalf()

        if left_limit * right_limit < 0: 
            inflection_points.append(point.evalf())

    return inflection_points

x = sp.Symbol('x')
f = x**2 - 30*x + 1
g = -x**3 - 7*x**2 + 5*x + 17
h = x**2 - 1/x
j = 1/x**5 + x**17 + 8

inflection_f = find_inflection_points(f)
inflection_g = find_inflection_points(g)
inflection_h = find_inflection_points(h)
inflection_j = find_inflection_points(j)

print("Inflection points for f(x):", inflection_f)
print("Inflection points for g(x):", inflection_g)
print("Inflection points for h(x):", inflection_h)
print("Inflection points for j(x):", inflection_j)
```

## Задание 8 (1.5 балла)

Реализуйте [метод Ньютона](https://en.wikipedia.org/wiki/Newton%27s_method_in_optimization) (метод второго порядка) для поиска минимума функции.

Про метод ещё можно почитать [здесь](https://math.semestr.ru/optim/minimum.php).

```python
# your code here

import sympy as sp

def newton_method(func, x_start, epsilon=1e-6, max_iterations=100, delta=1e-6):
    x = sp.Symbol('x')
    d_x = sp.diff(func, x)  
    d_xx = sp.diff(d_x, x)  

    x_curr = x_start
    trace = [x_curr]

    for i in range(max_iterations):
        grad = d_x.subs(x, x_curr)  
        hessian = d_xx.subs(x, x_curr)  

        if abs(grad) < delta:
            print(f"Градиент близок к 0 на {i}-й итерации.")
            break

        if hessian == 0:
            raise ValueError("Гессиан равен 0. Метод Ньютона не применим.")

        x_new = x_curr - grad / hessian  
        trace.append(x_new)

        if abs(x_new - x_curr) < epsilon:
            print(f"Метод сошелся на {i}-й итерации.")
            break

        x_curr = x_new

    return x_curr, trace
```

## Задание 9 (0.5 балла)

Примените метод Ньютона для поиска минимума указанных функций (в ответе выведите подходящее значение начальной точки, график с найденным и подписанным значением минимума, как в задании 7).

```python
# your code here

import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Определение функций
x = sp.Symbol('x')
definitions = {
    'f': x**2 - 30*x + 1,
    'g': -x**3 - 7*x**2 + 5*x + 17,
    'h': x**2 - 1/x,
    'j': 1/x**5 + x**17 + 8
}

# Метод Ньютона
def newton_method(func, x_start, learning_rate=0.1, epsilon=1e-6, delta=1e-6, max_iter=100):
    f_prime = sp.diff(func, x)
    f_double_prime = sp.diff(f_prime, x)
    x_curr = x_start

    trace = []
    for _ in range(max_iter):
        trace.append(float(x_curr))
        grad = f_prime.subs(x, x_curr)
        hess = f_double_prime.subs(x, x_curr)

        if abs(grad) < delta:
            break

        x_new = x_curr - learning_rate * (grad / hess)

        if abs(x_new - x_curr) < epsilon:
            x_curr = x_new
            break

        x_curr = x_new

    return float(x_curr), trace

# Построение графиков и применение метода Ньютона
for name, func in definitions.items():
    f_lambdified = sp.lambdify(x, func, modules=['numpy'])
    f_prime = sp.diff(func, x)

    x_start = 0 if name != 'j' else 0.5  # Для функции j избегаем x=0
    x_min, trace = newton_method(func, x_start)
    y_min = f_lambdified(x_min)

    # График
    x_vals = np.linspace(-10, 10, 1000) if name != 'j' else np.linspace(0.1, 2, 1000)
    y_vals = f_lambdified(x_vals)

    plt.figure(figsize=(8, 6))
    plt.plot(x_vals, y_vals, label=f'{name}(x)', color='blue')
    plt.scatter([x_min], [y_min], color='red', label=f'Minimum: ({x_min:.2f}, {y_min:.2f})')
    plt.annotate(f'({x_min:.2f}, {y_min:.2f})', (x_min, y_min), textcoords="offset points", xytext=(-15, 10), ha='center')

    plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
    plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
    plt.title(f'Graph of {name}(x) with Minimum')
    plt.xlabel('x')
    plt.ylabel(f'{name}(x)')
    plt.legend()
    plt.grid()
    plt.show()

    print(f"{name}: Minimum at x = {x_min:.6f}, y = {y_min:.6f}")
```
